```python
import numpy as np, matplotlib.pyplot as plt, seaborn as sns, pandas as pd
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df.groupby("work_year")["salary_in_usd"].mean().plot(kind="line", marker='o', title="Average Salary Over Years")
plt.xlabel("Work Year")
plt.ylabel("Average Salary (USD)")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
sns.countplot(data=df, x="experience_level")
plt.title("Experience Level Distribution")
plt.xlabel("Experience Level")
plt.ylabel("Count")
plt.show()
```

```python
top_jobs = df.groupby("job_title")["salary_in_usd"].mean().sort_values(ascending=False).head(10)
top_jobs.plot(kind="barh", title="Top 10 Highest Paying Job Titles")
plt.xlabel("Average Salary (USD)")
plt.gca().invert_yaxis()
plt.show()
```

```python
df.plot.scatter(x="remote_ratio", y="salary_in_usd",  title="Remote Ratio vs Salary")
plt.xlabel("Remote Ratio (%)")
plt.ylabel("Salary (USD)")
plt.show()
```

```python
df["employment_type"].value_counts().plot.pie(autopct="%1.1f%%", title="Employment Type Distribution")
plt.ylabel("")
plt.show()
```
